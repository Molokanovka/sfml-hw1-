```python
# Видим, что часть данных - object, скорее всего стоки.
# Давайте выведем эти значения для каждого столбца
```

```python
for i in all_data.columns: # перебираем все столбцы
    if str(all_data[i].dtype) == 'object': # если тип столбца - object
        print(i) # выводим название столбца
        print(set(all_data[i])) # выводим все его значения (но делаем set - чтоб значения не повторялись)
        print() # выводим пустую строку
```

```python
# можно заметить что некоторые переменные, 
# которые обозначены как строки (например PERSONAL_INCOME) на самом деле числа,
# но по какой-то (ниже увидим) причине были распознаны как строки
```

```python
all_data['PERSONAL_INCOME'] = all_data['PERSONAL_INCOME'].map(lambda x: x.replace(',', '.')).astype('float')
```

```python
# причина же что использовалась запятая для разделения не целой части числа
```

```python
all_data['CREDIT'] = all_data['CREDIT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['FST_PAYMENT'] = all_data['FST_PAYMENT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['LOAN_AVG_DLQ_AMT'] = all_data['LOAN_AVG_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
all_data['LOAN_MAX_DLQ_AMT'] = all_data['LOAN_MAX_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
```

```python
def NaN_GEN_INDUSTRY(s):
    if (s != s) :
        return 'Другие сферы' 
    else : 
        return s
        
```

```python
all_data['GEN_INDUSTRY']=all_data['GEN_INDUSTRY'].map(NaN_GEN_INDUSTRY)
```

```python
def NaN_GEN_TITLE(s):
    if (s != s) :
        return 'Начало карьеры' 
    else : 
        return s
```

```python
all_data['GEN_TITLE']=all_data['GEN_TITLE'].map(NaN_GEN_TITLE)
```

```python
def NaN_ORG_TP_STATE(s):
    if (s != s) :
        return 'Индивидуальный предприниматель' 
    else : 
        return s
```

```python
all_data['ORG_TP_STATE']=all_data['ORG_TP_STATE'].map(NaN_ORG_TP_STATE)
```

```python
def NaN_ORG_TP_FCAPITAL(s):
    if (s != s) :
        return 'С участием' 
    else : 
        return s
```

```python
all_data['ORG_TP_FCAPITAL']=all_data['ORG_TP_FCAPITAL'].map(NaN_ORG_TP_FCAPITAL)
```

```python
def NaN_JOB_DIR(s):
    if (s != s) :
        return 'Другое' 
    else : 
        return s
```

```python
all_data['JOB_DIR']=all_data['JOB_DIR'].map(NaN_JOB_DIR)
```

```python
def NaN_TP_PROVINCE(s):
    if (s != s) :
        return 'Не брал' 
    else : 
        return s
```

```python
all_data['TP_PROVINCE']=all_data['TP_PROVINCE'].map(NaN_TP_PROVINCE)
```

```python
def NaN_WORK_TIME(s):
    if (s != s) :
        return 0 
    else : 
        return s
```

```python
all_data['WORK_TIME']=all_data['WORK_TIME'].map(NaN_WORK_TIME)
```

```python
def NaN_PREVIOUS_CARD_NUM_UTILIZED(s):
    if (s != s) :
        return 0 
    else : 
        return s
```

```python
all_data['PREVIOUS_CARD_NUM_UTILIZED']= all_data['PREVIOUS_CARD_NUM_UTILIZED'].map(NaN_PREVIOUS_CARD_NUM_UTILIZED)
```
